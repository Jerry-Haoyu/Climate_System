### Data Reanaysis

- A process to genereate multi-decade, homogenized dat. 
- Hence mitigate the data bottleneck for climate studies 

> **Example**
> Suppose we have two temeperature measurements 
> $$
> \begin{cases}
> T_1&=T_t + \epsilon_1 \\
> T_2&=T_t + \epsilon_2
> \end{cases}
> $$
> We want a linear combination of these hence $T=aT_1 + (1-a)T_2$. Note this $$T\sim N(T_t, a^2\sigma_1^2 + (1-a)^2 \sigma_2^2)$$
> The variance is $a^2\sigma_1^2 + (1-2a+a^2)\sigma_2^2=a^2(\sigma_1^2 +\sigma_2^2)+(1-2a)\sigma_2^2$
> Taking the derivative we get:
> $$
> \frac{\partial \sigma}{\partial a}=2a(\sigma_1^2+\sigma_2^2)-2\sigma_2^2=0 \Rightarrow a=\frac{\sigma_2^2}{\sigma_1^2 + \sigma_2^2}
> $$
